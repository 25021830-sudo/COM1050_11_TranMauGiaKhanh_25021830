```python
#W4A2
def str_to_int(num: str) -> int:
    try:
        return int(num)
    except ValueError:
        return None

def check_positive_int(num):
    tmp = str_to_int(num)
    if tmp != None and tmp > 0:
        return tmp
    return 0

def is_prime(n: int):
    i = 2
    n = int(n)
    # print(type(i), type(n))
    while(int(i*i) <= n):
        if(n % i == 0):
            return False
        i = i + 1
    return True

n = input()
while(check_positive_int(n) == 0):
    n = input()

print(is_prime(n))
```

```python
#W4A8
def is_palindrome(num) :
    num = str(num)
    length = len(num)
    for i in range (0, length//2):
        if num[i]!=num[length-i-1]:
            return False
    return True

def rev(num) :
    num = str(num)
    length = len(num)
    revnum = ""
    for i in range(0, length):
        revnum=num[i]+revnum
    return int(revnum)

n = int(input())
step=0
while not is_palindrome(n):
    n = n + rev(n)
    step+=1

print(step, n)
```

```python
#W4A10
def collatz(num) :
    count=1
    while(num!=1):
        if num&1==0:
            num//=2
            count+=1
        else :
            num=3*num+1
            count+=1
    return count

maxnum = -999
maxlen = -999
n = int(input())
for i in range (1, n+1) :
    curlen = collatz(i)
    if curlen>maxlen:
        maxnum=i
        maxlen=curlen

print(maxnum, maxlen)
```
