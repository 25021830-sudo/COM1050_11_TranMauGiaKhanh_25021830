```python
#W5A1
def max_of_two(a,b):
    if a == b:
        return a
    if a - b > 0 :
        return a
    else:
        return b
a,b = map(int,input().split())
print(max_of_two(a,b))
```

```python
#W5A3
def IsPrime(a):
    if a == 0 or a == 1 or a == 2:
        return "False"
    if a > 1 :
        for i in range(2,a,1):
            if a % i == 0:
                return "False"
        return "True"
a = int(input())
print(IsPrime(a))
```

```python
#W5A4
def uoc(a):
    ds = []
    i = 0
    while i < a - 1 :
        i += 1
        if a % i == 0:
            ds.append(i)
    return ds
def perfect_number(a):
    b = sum(uoc(a))
    if a == b:
        return True
    else:
        return False
a = int(input())
print(b)
print(a)

print(perfect_number(a))
```

```python
#W5A5
def ham_chi_so(lst2,k):
    for i in range(0,len((lst2))+1):
        if lst2[i] == k:
                return i + 1
    return -1
lst = map(int,input().split())
lst2 = list(lst)
k = int(input())
print(ham_chi_so(lst2,k))
```

```python
#W5A6
def factorial(a):
    i = 1
    result = 1
    if a == 0 or a == 1:
        return 1
    if a > 1:
        while i < a:
            i += 1
            result *= i
    return result
a = int(input())
print(factorial(a))
```

```python
#W5A7
def cal(n1,n2,op):
    n1 = float(n1)
    n2 = float(n2)
    if op == "*":
        return n1 * n2
    elif op == "/":
        return n1 / n2
    elif op == "+":
        return n1 + n2
    elif op == "-":
        return n1 - n2
n1,op,n2 = map(str,input().split())
print(cal(n1,n2,op))
```

```python
#W5A9
def tach_so(n):
    ds = []
    for i in range(0,len(str(n)),1):
        b = n // 10**(i) - n // 10**(i+1) * 10
        ds.append(b)
    return ds

def tong(ds):
    return sum(list(ds))

ds = tach_so(n)

n = int(input())
print(tong(ds))
```

```python
#W5A10
def anh_xa(strA,strB):
    k = {}
    l = {}


    for i in range(0,len(strA)):
        chA = strA[i]
        chB = strB[i]
        if chA not in k:
            k[chA] = chB
        else:
            if chA != chB:
                return False
    for i in range(0,len(strA)):
        chA = strA[i]
        chB = strB[i]
        if chB not in k:
            l[chB] = chA
        else:
            if chB != chA:
                return False
    return True
strA,strB = input().split()
print(anh_xa(strA,strB))
```
